### 2.6 用于多分类的线性模型

```python
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs
import mglearn


X, y = make_blobs(random_state=42)
mglearn.discrete_scatter(X[:, 0], X[:, 1], y)
plt.xlabel("Feature 0")
plt.ylabel("Feature 1")
plt.legend(["Class 0", "Class 1", "Class 2"])
```

```python
from sklearn.svm import LinearSVC


linear_svm = LinearSVC().fit(X, y)
print(f"Coefficient shape: {linear_svm.coef_.shape}")
print(f"Intercept shape: {linear_svm.intercept_.shape}")
```

```python
linear_svm.intercept_
```

```python
import numpy as np


mglearn.discrete_scatter(X[:, 0], X[:, 1], y) # 输出数据集的可视化效果，横轴为第一个特征，纵轴为第二个特征，用y分类
line = np.linspace(-15, 15)

for coef, intercept, color in zip(linear_svm.coef_, linear_svm.intercept_,["b", "r", "g"]):
    plt.plot(line, -(line * coef[0] + intercept) / coef[1], color=color)
plt.ylim(-10, 15)
plt.xlim(-10, 8)
plt.xlabel("Feature 0")
plt.ylabel("Feature 1")
plt.legend(["Class 0", "Class 1", "Class 2", "Line class 0", "Line class 1", "Line class 2"], loc="upper left")
```
