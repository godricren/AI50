## 强制类型统一
numpy设计初衷是用于运算的，所以对数据类型进行统一优化<br>
**注意**：
1. numpy默认ndarray的所有元素类型是相同的
2. 如果传进来的列表中包含不同的类型，则强制统一为同一类型，优先级为**str>float>int**
